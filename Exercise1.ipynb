## C. Cleaning



Check for missing values

```python
df.isnull().sum()
```

Check for outliers in `minimum_nights`

```python
df['minimum_nights'].describe()
```

Check for invalid values (e.g., price ≤ 0)

```python
invalid_latitude_count = df[(df['latitude'] < -90) | (df['latitude'] > 90)].shape[0]
print(f"Invalid latitude values: {invalid_latitude_count}")

invalid_longitude_count = df[(df['longitude'] < -180) | (df['longitude'] > 180)].shape[0]
print(f"Invalid longitude values: {invalid_longitude_count}")

invalid_price_count = df[df['price'] <= 0].shape[0]
print(f"Price values less than or equal to 0: {invalid_price_count}")

invalid_minimum_nights_count = df[df['minimum_nights'] <= 0].shape[0]
print(f"Minimum nights values less than or equal to 0: {invalid_minimum_nights_count}")

invalid_reviews_per_month_count = df[df['reviews_per_month'] < 0].shape[0]
print(f"Reviews per month values less than 0: {invalid_reviews_per_month_count}")

invalid_availability_count = df[(df['availability_365'] < 0) | (df['availability_365'] > 365)].shape[0]
print(f"Availability_365 values outside 0-365 range: {invalid_availability_count}")
```

Apply a missing-value strategy (drop, fill, or flag)


```python
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
df['last_review'] = df['last_review'].fillna('N/A')
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

print(df.isnull().sum())
```

Apply an outlier-handling strategy

```python
min_nights_99th_percentile = df['minimum_nights'].quantile(0.99)
df['minimum_nights'] = np.where(df['minimum_nights'] > min_nights_99th_percentile, min_nights_99th_percentile, df['minimum_nights'])

print(f"Minimum nights capped at: {min_nights_99th_percentile}")
print(df['minimum_nights'].describe())
```

## D. Enriching

Create at least 3 new variables:
- price_per_minimum_night
- review_rate_per_month (combine ratings + frequency)
- is_expensive (boolean flag based on price threshold)


```python
df['price_per_minimum_night'] = df['price'] / df['minimum_nights']
df['review_rate_per_month'] = df['number_of_reviews'] / df['reviews_per_month']
price_75th_percentile = df['price'].quantile(0.75)
df['is_expensive'] = df['price'] > price_75th_percentile
```

Create 1 aggregated summary table

(e.g. average price per neighborhood, number of hosts per borough)

```python
aggregated_table = df.groupby('neighbourhood_group').agg(
    average_price=('price', 'mean'),
    number_of_hosts=('host_id', 'nunique')
).reset_index()

print(aggregated_table)
```

## E. Validating

Create at least 3 validation checks
- Price must be positive.
- Host ID should not be null.
- Minimum nights must be ≥ 1.
- Generate a validation report:
- Count of rows violating each rule.

Decide if invalid rows (if any) should be removed or corrected?

```python
invalid_price_count = df[df['price'] <= 0].shape[0]
print(f"Price values less than or equal to 0: {invalid_price_count}")

invalid_minimum_nights_count = df[df['minimum_nights'] < 1].shape[0]
print(f"Minimum nights values less than or equal to 1: {invalid_minimum_nights_count}")

invalid_host_id_count = df[df['host_id'].isnull()].shape[0]
print(f"Invalid host_id values: {invalid_host_id_count}")

min_valid_price = df.loc[df['price'] > 0, 'price'].min()
print(f"Minimum valid price: {min_valid_price}")
df.loc[df['price'] <= 0, 'price'] = min_valid_price

invalid_price_count = df[df['price'] <= 0].shape[0]
print(f"Price values less than or equal to 0: {invalid_price_count}")
```
